```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('anscombe.csv')
print(df.head())
```

SCATTER PLOT

```python

sns.set_style('whitegrid')
scatter_plot = sns.relplot(data=df, x='x', y='y', hue='dataset', kind='scatter', height=6, aspect=1.5)
scatter_plot.set(title='Scatter Plot', xlabel='X-axis', ylabel='Y-axis')
```

 Line Plot

```python

line_plot = sns.relplot(data=df, x='x', y='y', hue='dataset', kind='line', marker='o', height=6, aspect=1.5)
line_plot.set(title='Line Plot', xlabel='X-axis', ylabel='Y-axis')
```

Bar Plot

```python

bar_data = df.groupby('dataset')['y'].mean().reset_index()
bar_plot = sns.catplot(data=bar_data, x='dataset', y='y', kind='bar', palette='viridis', height=6, aspect=1.5)
bar_plot.set(title='Bar Plot', xlabel='Dataset', ylabel='Mean of Y')
```

Count Plot

```python

count_plot = sns.catplot(data=df, x='dataset', kind='count', palette='viridis', height=6, aspect=1.5)
count_plot.set(title='Count Plot', xlabel='Dataset', ylabel='Count')
```

 Box Plot

```python

box_plot = sns.catplot(data=df, x='dataset', y='y', kind='box', palette='Set2', height=6, aspect=1.5)
box_plot.set(title='Box Plot', xlabel='Dataset', ylabel='Y')
```
