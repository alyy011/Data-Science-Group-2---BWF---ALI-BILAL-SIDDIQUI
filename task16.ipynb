```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('batter.csv')
print(df.head())
```

```python
#Simple line plot
x = df['batter']
y = df['runs']

plt.figure()
plt.plot(x, y)
plt.title('Simple Line Plot')
plt.xlabel('batter')
plt.ylabel('runs')
plt.show()
```

```python
#Plot multiple lines on the same graph.
x = df['batter']
y1 = df['runs']
y2 = df['strike_rate']  

plt.figure()
plt.plot(x, y1, label='Runs')
plt.plot(x, y2, label='Average')
plt.title('Multiple Lines on the Same Graph')
plt.xlabel('Batter')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
#Add markers to a line plot
plt.figure()
plt.plot(x, y1, marker='o')
plt.title('Line Plot with Markers')
plt.xlabel('Batter')
plt.ylabel('Runs')
plt.show()
```

```python
#Customize line styles and colors
plt.figure()
plt.scatter(x, y1, marker='x', s=100, c='r')
plt.title('Customized Scatter Plot')
plt.xlabel('Batter')
plt.ylabel('Runs')
plt.show()
```

```python
#Plot multiple scatter plots on the same graph
plt.figure()
plt.scatter(x, y1, marker='o', s=50, c='r', label='Runs')
plt.scatter(x, y2, marker='x', s=100, c='b', label='Average')
plt.title('Multiple Scatter Plots')
plt.xlabel('Batter')
plt.ylabel('Values')
plt.legend()
plt.show()
```

```python
#Add labels and titles to the scatter plot
plt.figure()
plt.scatter(x, y1, marker='o', s=50, c='r')
plt.title('Scatter Plot with Labels')
plt.xlabel('Batter')
plt.ylabel('Runs')
plt.show()
```

Bar Charts


```python
#Create a vertical bar chart
plt.figure()
plt.bar(x, y1, color='blue', edgecolor='black')
plt.title('Vertical Bar Chart')
plt.xlabel('Batter')
plt.ylabel('Runs')
plt.show()
```

```python
#Create a horizontal bar chart
plt.figure()
plt.barh(x, y1, color='green', edgecolor='black')
plt.title('Horizontal Bar Chart')
plt.xlabel('Runs')
plt.ylabel('Batter')
plt.show()
```

```python
#Customize bar colors and add edge colors
plt.figure()
plt.bar(x, y1, color='purple', edgecolor='yellow')
plt.title('Customized Bar Colors')
plt.xlabel('Batter')
plt.ylabel('Runs')
plt.show()
```
