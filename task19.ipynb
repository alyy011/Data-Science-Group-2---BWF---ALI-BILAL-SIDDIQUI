```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('anscombe.csv')
print(df.head())
```

```python

sns.set_style('whitegrid')
```

```python
# Linear Regression Plot
sns.regplot(data=df, x='x', y='y')
plt.title('Linear Regression Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Multiple Linear Regression Plot
sns.lmplot(data=df, x='x', y='y', hue='dataset', height=6, aspect=1.5)
plt.title('Multiple Linear Regression Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Residual Plot
sns.residplot(data=df, x='x', y='y')
plt.title('Residual Plot')
plt.xlabel('X-axis')
plt.ylabel('Residuals')
plt.show()
```

```python
# Histogram
sns.histplot(data=df, x='y', bins=10, kde=True)
plt.title('Histogram')
plt.xlabel('Y-axis')
plt.ylabel('Frequency')
plt.show()
```

```python
# KDE Plot
sns.kdeplot(data=df, x='y', hue='dataset', fill=True)
plt.title('Kernel Density Estimate (KDE) Plot')
plt.xlabel('Y-axis')
plt.ylabel('Density')
plt.show()
```

```python
# Distribution Plot
sns.displot(data=df, x='y', hue='dataset', kind='kde', fill=True)
plt.title('Distribution Plot')
plt.xlabel('Y-axis')
plt.ylabel('Density')
plt.show()
```

```python
# Box Plot
sns.boxplot(data=df, x='dataset', y='y')
plt.title('Box Plot')
plt.xlabel('Dataset')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Violin Plot
sns.violinplot(data=df, x='dataset', y='y')
plt.title('Violin Plot')
plt.xlabel('Dataset')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Pair Plot
sns.pairplot(df, hue='dataset')
plt.title('Pair Plot')
plt.show()
```

```python
# Heatmap (Using a correlation matrix for demonstration)
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap')
plt.show()
```

```python
# Joint Plot
sns.jointplot(data=df, x='x', y='y', kind='scatter')
plt.title('Joint Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# Facet Grid
g = sns.FacetGrid(df, col='dataset')
g.map(sns.scatterplot, 'x', 'y')
g.add_legend()
plt.show()
```
